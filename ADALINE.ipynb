```python
import numpy as np
import matplotlib.pyplot as plt
import random
import pandas as pd
np.random.seed(0)
n_samples = 250
# Вариант 8
X = np.random.rand(n_samples, 2)
Y = (((X[:, 0]))+(0.5*X[:, 1]-0.3) < 0.4)
w = np.array([0,0])
b = 0
```

```python
def ADALINE(X,Y,n_samples,W,b,alpha):
    plt.scatter(X[Y,0],X[Y,1])
    plt.scatter(X[~Y,0],X[~Y,1],color = 'r')
    targ = []
    epochs =0
    MSE = 0
    for y in Y:
        if y == True:
            targ.append(1)
        else:
            targ.append(-1)
    while(epochs<10):
        for i in range(X.shape[0]):
            y_s = np.dot(W,X[i])+b
            error1 =(targ[i]-y_s)**2
            error =2*(targ[i]-y_s)

            MSE+=error1
            W = W + alpha*error *X[i]
            b = b + alpha*error
        epochs+=1
        if(MSE>10000):
            
            break
        
    X = np.array([0,1])
    Y = -((W[0]/W[1]))*X - b/W[1]
    print("W = ", W )
    print("b = ", b)
    print("MSE = ", MSE)
    plt.plot(X,Y)
    plt.grid()       
    
    
        
#eps=max((np.linalg.eig(1/2*(X[:,0].reshape(-1,1)@X[:,0].reshape(1,-1))+(X[:,1].reshape(-1,1)@X[:,1].reshape(1,-1))))[0])
# R = 1/2*(X[:,0].reshape(-1,1)@X[:,0].reshape(1,-1))+(X[:,1].reshape(-1,1)@X[:,1].reshape(1,-1))[0]
# print(R.shape)
# R = np.zeros((2,2))
# for i in range(len(X)):
#     R += X[i].reshape(2,1)@X[i].reshape(1,2)
# R = R*1/2    
# print(R)  
# print(np.linalg.eig(R))
alpha = 0.05
ADALINE(X,Y,n_samples,w,b,alpha)    
```
