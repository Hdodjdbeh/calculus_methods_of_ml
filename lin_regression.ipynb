```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(0)
```

### ADALINE

```python
#  8
x_objects = np.random.uniform(0, 1, size=500)
y_objects = (-3 * x_objects)*np.exp(x_objects) + np.random.normal(scale=0.7, size=x_objects.shape)
plt.scatter(x_objects, y_objects, alpha = 0.3)#  7
```

```python
def ADALINE(X,Y,n_samples,W,b,alpha):
    epochs =0
    MSE = 0
    while(epochs<20):
        for i in range(X.shape[0]):
            y_s = X[i]*W+b
            error1 =(Y[i]-y_s)**2
            grad_error =2*(Y[i]-y_s)

            MSE+=error1
            W = W + alpha*grad_error*X[i]
            b = b + alpha*grad_error
        epochs+=1
        if(MSE<0.1):
            break



    #Y = -((W[0]/W[1]))*X - b/W[1]
    print("W = ", W )
    print("b = ", b)
    #print("MSE = ", MSE)

    return W,b

w=0
b=0
alpha = 0.05
w,b=ADALINE(x_objects,y_objects,x_objects.shape[0],w,b,alpha)


plt.scatter(x_objects, y_objects, alpha = 0.3)
x=np.array([0,1])
y=w*x+b
plt.plot(x,y)

```

```python
#  8
x = np.random.uniform(0, 1, size=500)
y = (-3 * x)*np.exp(x) + np.random.normal(scale=0.7, size=x.shape)





C=np.cov(x,y)
a=C[0,1]/C[0,0]
b=np.mean(y)-a*np.mean(x)
xfit = np.array([0,1])
yfit = a*xfit + b
print(f"Model coef: a={a}")
print(f"Model intercept: b={b}")
plt.figure()
plt.scatter(x, y, alpha = 0.3)
plt.plot(xfit, yfit)
```

### Sklearn

```python
#  8
#  8
x = np.random.uniform(0, 1, size=500)
y = (-3 * x)*np.exp(x) + np.random.normal(scale=0.7, size=x.shape)

from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)
model.fit(x[:, np.newaxis], y)
xfit = np.array([0,1])
yfit = model.predict(xfit[:, np.newaxis])
plt.figure()
plt.scatter(x, y, alpha = 0.3)
plt.plot(xfit, yfit)

print(f"Model coef: a={model.coef_[0]}")
print(f"Model intercept: b={model.intercept_}")
```
